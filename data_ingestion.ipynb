```python
import pandas as pd
import numpy as np
```

```python
data =  pd.read_csv("C:\\Users\\Advance solution\\Downloads\\31. Laptop_CSV_File.csv")
```

```python
df = data.copy()
```

```python
df = df.iloc[:, 1:]
```

```python
df['Ram'] = df['Ram'].str.replace('GB','')
```

```python
df['Ram'] = df['Ram'].astype('int32')
```

```python
df['Weight'] = df['Weight'].str.replace('kg','')
```

```python
df['Weight'] = df['Weight'].astype('float32')
```

```python
df.to_csv("C:\\Users\\Advance solution\\OneDrive\\Desktop\\New\\first_clean.csv", index=False)
```
